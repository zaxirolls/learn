```python
hier_df = DataFrame(
    np.arange(9).reshape((3,3)),
    index = [
        ['a','a','b'],
        [1,2,2]
    ],
    columns = [
        ['osaka','Tokyo','Osaka'],
        ['Blue','Red','Red']
    ]
)
hier_df
```

```python
hier_df.index.names = ['key1','key2']
hier_df.columns.names = ['city','color']
hier_df
```
