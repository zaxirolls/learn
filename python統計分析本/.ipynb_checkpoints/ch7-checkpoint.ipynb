```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
%matplotlib inline

icecream = [[1,464],[2,397],[3,493],[4,617],[5,890],[6,883],[7,1292],[8,1387],[9,843],[10,621],[11,459],[12,561]]

temperature = [[1,10.6],[2,12.2],[3,14.9],[4,20.3],[5,25.2],[6,26.3],[7,29.7],[8,31.6],[9,27.7],[10,22.6],[11,15.5],[12,13.8]]

x = np.array([u[1] for u in temperature])
y = np.array([u[1] for u in icecream])

result = scipy.stats.linregress(x,y)
print('傾き=',result.slope.round(4),'切片=',result.intercept.round(4),'信頼係数=',result.rvalue.round(4),'p値=',result.pvalue.round(4),'標準誤差=',result.stderr.round(4))

b = result.slope
a = result.intercept
plt.plot(x,[b*u + a for u in x])
plt.scatter(x,y)
plt.title('2016年の気温と一世帯あたりアイスクリーム支出')
plt.xlabel('月間平均気温')
plt.ylabel('月間アイスクリーム支出')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
from sklearn import linear_model
%matplotlib inline

icecream = [[1,464],[2,397],[3,493],[4,617],[5,890],[6,883],[7,1292],[8,1387],[9,843],[10,621],[11,459],[12,561]]

temperature = [[1,10.6],[2,12.2],[3,14.9],[4,20.3],[5,25.2],[6,26.3],[7,29.7],[8,31.6],[9,27.7],[10,22.6],[11,15.5],[12,13.8]]

x = pd.DataFrame([u[1] for u in temperature])
y = pd.DataFrame([u[1] for u in icecream])

model = linear_model.LinearRegression()
results = model.fit(x,y)

print('a',model.coef_[0][0],'b',model.intercept_[0])

plt.plot(x,model.predict(x))
plt.scatter(x,y)
plt.title('2016年の気温と一世帯あたりアイスクリーム支出')
plt.xlabel('月間平均気温')
plt.ylabel('月間アイスクリーム支出')
plt.show()
```
